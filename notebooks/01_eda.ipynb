```python
import pandas as pd
import numpy as np

clinical_lihc_pd = pd.read_csv('../data/raw/TCGA-LIHC.clinical.tsv', sep='\t')
methylation_lihc_pd = pd.read_csv('../data/raw/TCGA-LIHC.methylation450.tsv', sep='\t')
```

```python
#transpose the methylation dataframe

methylation_lihc_pd.set_index('Composite Element REF', inplace=True)
methylation_lihc_pd_old = methylation_lihc_pd.transpose()
methylation_lihc_pd_old
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# --- STEP 1: Drop probes with too many NaNs ---
# Calculate the threshold for 10% missingness
# (e.g., if you have 370 tumor samples, keep probes that have at least 333 valid values)
threshold = int(methylation_lihc_pd_old.shape[0] * 0.90)

# Drop columns (probes) that do not meet the threshold
methylation_lihc_pd = methylation_lihc_pd_old.dropna(axis=1, thresh=threshold)

print(f"Original probes: {methylation_lihc_pd_old.shape[1]}")
print(f"Probes after dropping bad ones: {methylation_lihc_pd.shape[1]}")

```

```python
methylation_lihc_pd
```

```python
# --- STEP 2: Impute remaining NaNs ---
# Initialize the imputer to fill NaNs with the median value of each column
imputer = SimpleImputer(strategy='median')

# Fit and transform the data (Note: this returns a numpy array)
methylation_imputed_array = imputer.fit_transform(methylation_lihc_pd)

# Convert back to a pandas DataFrame to keep your Patient IDs and Probe IDs safe
methylation_lihc_pd = pd.DataFrame(
    methylation_imputed_array,
    index=methylation_lihc_pd.index,      # Patient IDs
    columns=methylation_lihc_pd.columns   # CpG Probe IDs
)

print(f"Total NaNs remaining: {methylation_lihc_pd.isna().sum().sum()}")
```

```python
columns_to_keep = [
    # --- Identifiers & Subsetting ---
    'sample',                                  # Keep as index to match methylation matrix
    'sample_type.samples',                     # Essential: For subsetting (Tumor vs Normal)
    'tissue_type.samples',                     # Alternative/backup for subsetting
    
    # --- Disease Scope (Useful if merging with other TCGA cohorts later) ---
    'disease_type',                            # e.g., Liver Hepatocellular Carcinoma
    'primary_site',                            # e.g., Liver
    
    # --- Clinical & Progression ---
    'tumor_grade.diagnoses',                   # Highly correlated with methylation changes
    'ajcc_pathologic_stage.diagnoses',         # Overall disease stage
    
    # --- Risk Factors & Demographics ---
    'alcohol_history.exposures',               # Major epigenetic driver in LIHC
    'gender.demographic',                      # Strong bias in LIHC / Check for sex chromosome filtering
    'race.demographic',                        # Check for disparities/patterns
    'vital_status.demographic',                # Survival correlation
    
    # --- Technical / Batch Effect Checking ---
    'tissue_source_site_id.tissue_source_site',# ID of the collection hospital
    'name.tissue_source_site'                  # Name of the collection hospital
]

# Filter your phenotype dataframe
clinical_lihc_pd_filtered = clinical_lihc_pd[columns_to_keep]
clinical_lihc_pd_filtered
```

```python
# Shows the count of unique classes per column across the whole dataset
unique_counts = clinical_lihc_pd_filtered.nunique()
print(unique_counts)
```

```python
clinical_lihc_pd_filtered = clinical_lihc_pd_filtered.drop(columns=["disease_type", "primary_site", "tumor_grade.diagnoses", "alcohol_history.exposures"])
clinical_lihc_pd_filtered
```

```python
clinical_lihc_pd_filtered_all = clinical_lihc_pd_filtered.copy()
clinical_lihc_pd_filtered_tumor = clinical_lihc_pd_filtered[clinical_lihc_pd_filtered['sample_type.samples'] == 'Primary Tumor']
```

```python
# 1. Extract the valid sample IDs from your clinical dataframe into a list
valid_tumor_samples = clinical_lihc_pd_filtered_tumor['sample'].tolist()
valid_all_samples = clinical_lihc_pd_filtered_all['sample'].tolist() # Assuming you have this too

# 2. Filter the methylation dataframe keeping only rows where the Index matches the list
methylation_lihc_pd_tumor = methylation_lihc_pd[methylation_lihc_pd.index.isin(valid_tumor_samples)]
methylation_lihc_pd_all = methylation_lihc_pd[methylation_lihc_pd.index.isin(valid_all_samples)]
```
