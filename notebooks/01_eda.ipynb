```python
import pandas as pd
import numpy as np

clinical_hnsc_pd = pd.read_csv('../data/raw/TCGA-HNSC.clinical.tsv', sep='\t')
methylation_hnsc_pd = pd.read_csv('../data/raw/TCGA-HNSC.methylation450.tsv', sep='\t')
```

```python
#transpose the methylation dataframe

methylation_hnsc_pd.set_index('Composite Element REF', inplace=True)
methylation_hnsc_pd_old = methylation_hnsc_pd.transpose()
methylation_hnsc_pd_old
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# --- STEP 1: Drop probes with too many NaNs ---
# Calculate the threshold for 10% missingness
# (e.g., if you have 370 tumor samples, keep probes that have at least 333 valid values)
threshold = int(methylation_hnsc_pd_old.shape[0] * 0.90)

# Drop columns (probes) that do not meet the threshold
methylation_hnsc_pd = methylation_hnsc_pd_old.dropna(axis=1, thresh=threshold)

print(f"Original probes: {methylation_hnsc_pd_old.shape[1]}")
print(f"Probes after dropping bad ones: {methylation_hnsc_pd.shape[1]}")

```

```python
methylation_hnsc_pd
```

```python
print(f"Total NaNs remaining: {methylation_hnsc_pd.isna().sum().sum()}")
```

```python
# --- STEP 2: Impute remaining NaNs ---
# Initialize the imputer to fill NaNs with the median value of each column
imputer = SimpleImputer(strategy='median')

# Fit and transform the data (Note: this returns a numpy array)
methylation_imputed_array = imputer.fit_transform(methylation_hnsc_pd)

# Convert back to a pandas DataFrame to keep your Patient IDs and Probe IDs safe
methylation_hnsc_pd = pd.DataFrame(
    methylation_imputed_array,
    index=methylation_hnsc_pd.index,      # Patient IDs
    columns=methylation_hnsc_pd.columns   # CpG Probe IDs
)

print(f"Total NaNs remaining: {methylation_hnsc_pd.isna().sum().sum()}")
```

```python
columns_to_keep = [
    # --- Identifiers & Subsetting ---
    'sample',                                  # Keep as index to match methylation matrix
    #'sample_type.samples',                     # Essential: For subsetting (Tumor vs Normal)
    'tissue_type.samples',                     # Alternative/backup for subsetting
    
    # --- Disease Scope (Useful if merging with other TCGA cohorts later) ---
    #'disease_type',                            # e.g., Liver Hepatocellular Carcinoma
    #'primary_site',                            # e.g., Liver
    
    # --- Clinical & Progression ---
    #'tumor_grade.diagnoses',                   # Highly correlated with methylation changes
    #'ajcc_pathologic_stage.diagnoses',         # Overall disease stage
    
    # --- Risk Factors & Demographics ---
    'alcohol_history.exposures'               # Major epigenetic driver in HNSC
    #'gender.demographic',                      # Strong bias in HNSC / Check for sex chromosome filtering
    #'race.demographic',                        # Check for disparities/patterns
    #'vital_status.demographic',                # Survival correlation
    
    # --- Technical / Batch Effect Checking ---
    #'tissue_source_site_id.tissue_source_site',# ID of the collection hospital
    #'name.tissue_source_site',                  # Name of the collection hospital
    
    #'cigarettes_per_day.exposures',               # Another major epigenetic driver in HNSC
    #'pack_years_smoked.exposures',               # Cumulative smoking exposure
    #'age_at_index.demographic',                      # Age at diagnosis (important for methylation patterns)
    #'tissue_or_organ_of_origin.diagnoses',                   # Tissue of origin (e.g., Oral Cavity, Larynx)
    #'primary_diagnosis.diagnoses',                   # Primary diagnosis (e.g., Squamous Cell Carcinoma)
]

# Filter your phenotype dataframe
clinical_hnsc_pd_filtered = clinical_hnsc_pd[columns_to_keep]
clinical_hnsc_pd_filtered
```
