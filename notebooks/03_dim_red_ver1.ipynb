```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA, KernelPCA
import umap

# Set plot style
sns.set_theme(style="whitegrid")

# 1. Load the Data
print("Loading data...")
df_meth = pd.read_csv('../data/processed/methylation_lihc_pd_all.csv', index_col=0)
df_clin = pd.read_csv('../data/processed/clinical_lihc_pd_filtered_all.csv')

# 2. Align the Metadata
print("Aligning clinical data with methylation data...")
df_clin.set_index('sample', inplace=True)
df_clin = df_clin.reindex(df_meth.index)

# Define the list of columns you want to visualize
columns_to_plot = [
    'sample_type.samples', 
    'tissue_type.samples', 
    'ajcc_pathologic_stage.diagnoses', 
    'gender.demographic', 
    'race.demographic', 
    'vital_status.demographic'
]

# Fill missing values for all target columns to prevent plotting errors
for col in columns_to_plot:
    df_clin[col] = df_clin[col].fillna('Unknown')

# 3. Preprocessing
print("Imputing missing values...")
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(df_meth)

# 4. Dimensionality Reduction (Run only ONCE)
print("Running PCA...")
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("Running Kernel PCA...")
kpca = KernelPCA(n_components=2, kernel='rbf', fit_inverse_transform=False, n_jobs=-1)
X_kpca = kpca.fit_transform(X)

print("Running UMAP...")
reducer = umap.UMAP(n_components=2, random_state=42, n_jobs=-1)
X_umap = reducer.fit_transform(X)

# 5. Visualization Loop
print("Plotting results...")

# Loop through each clinical column and generate a 1x3 figure
for col in columns_to_plot:
    color_labels = df_clin[col]
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    # Add a main title for the figure to know which variable we are looking at
    fig.suptitle(f'Colored by: {col}', fontsize=16, fontweight='bold')
    
    # We remove the explicit "Set1" palette and let Seaborn auto-assign colors 
    # (tab10) because different columns will have varying numbers of categories.
    
    # Plot PCA
    sns.scatterplot(
        x=X_pca[:, 0], y=X_pca[:, 1], 
        hue=color_labels, alpha=0.8, s=30, ax=axes[0]
    )
    axes[0].set_title('PCA')
    axes[0].set_xlabel('Principal Component 1')
    axes[0].set_ylabel('Principal Component 2')
    
    # Move the legend outside the PCA plot so it doesn't cover data points
    sns.move_legend(axes[0], "upper left", bbox_to_anchor=(1.05, 1))

    # Plot Kernel PCA
    sns.scatterplot(
        x=X_kpca[:, 0], y=X_kpca[:, 1], 
        hue=color_labels, alpha=0.8, s=30, ax=axes[1], legend=False 
    )
    axes[1].set_title('Kernel PCA (RBF)')
    axes[1].set_xlabel('Component 1')
    axes[1].set_ylabel('Component 2')

    # Plot UMAP
    sns.scatterplot(
        x=X_umap[:, 0], y=X_umap[:, 1], 
        hue=color_labels, alpha=0.8, s=30, ax=axes[2], legend=False 
    )
    axes[2].set_title('UMAP')
    axes[2].set_xlabel('UMAP 1')
    axes[2].set_ylabel('UMAP 2')

    plt.tight_layout()
    plt.show()
```
